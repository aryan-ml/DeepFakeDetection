# Spliting Code
* Video Wise Split
* Extract Unique Videos
* 70% Train, 15% Val, 15% Test
* Add 'split' column back into main dataframe
* Save Updated Manifest


```python
import pandas as pd
import os, math, random
from pathlib import Path
from sklearn.model_selection import train_test_split
```

```python
BASE_DIR = Path().resolve().parents[0]

csv_path = BASE_DIR / 'data' / 'processed' / 'manifest.csv'

df = pd.read_csv(csv_path)
df.head()
```

## Extract Unique Videos

```python
video_df = df[['video_id', 'label']].drop_duplicates()

video_df.head()
```

## 70% Train, 15% Val, 15% Test


```python
# Train split
train_vids, temp_vids = train_test_split(
    video_df,
    test_size=0.30,
    stratify=video_df['label'],
    random_state=1
)

# Validation + Test split
val_vids, test_vids = train_test_split(
    temp_vids,
    test_size=0.50,
    stratify=temp_vids['label'],
    random_state=1
)
```

## Add 'split' column back into main dataframe

```python
df['split'] = None
df.loc[df['video_id'].isin(train_vids['video_id']), 'split'] = 'train'
df.loc[df['video_id'].isin(val_vids['video_id']), 'split'] = 'val'
df.loc[df['video_id'].isin(test_vids['video_id']), 'split'] = 'test'

df.split.value_counts()
```

## Save Updated Manifest

```python
df.to_csv(csv_path, index=False)
print("Updated manifest saved.")
```
